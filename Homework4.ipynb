```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('dsu1.jpg')
rows,cols,ch = img1.shape

pts1 = np.float32([[325,280],[763,312],[324,537],[766,509]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img1,M,(300,300))

plt.subplot(121),plt.imshow(img1),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
img2 = cv2.imread('dsu2.jpg')
rows,cols,ch = img2.shape

pts3 = np.float32([[180,242],[823,220],[175,503],[819,511]])
pts4 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M2 = cv2.getPerspectiveTransform(pts3,pts4)

dst2 = cv2.warpPerspective(img2,M2,(300,300))

plt.subplot(121),plt.imshow(img2),plt.title('Input')
plt.subplot(122),plt.imshow(dst2),plt.title('Output')
plt.show()
```

```python
addh = cv2.hconcat([dst,dst2])

cv2.imshow('imgh',addh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img3 = cv2.add(dst,dst2)

cv2.imshow('img',img3)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('hconcatdsu.jpg',addh)
cv2.imwrite('adddsu.jpg',img3)
```

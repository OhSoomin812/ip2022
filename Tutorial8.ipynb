```python
class Reverse:
  "Iterator for looping over a sequence backwards"
  def __init__(self, data):
    self.data = data
    self.index = len(data)
  def next(self):
    if self.index == 0:
      raise StopIteration
    self.index = self.index - 1
    return self.data[self.index]
  def __iter__(self):
    return self
```

```python
for char in Reverse('spam') : 
    print (char)
```

```python
[x for x in Reverse('spam')]
```

```python
[x + x for x in Reverse('spam')]
```

```python
def reverse(data):
  for i in range(len(data)):
    yield data[len(data)-1-i]
```

```python
for char in reverse('spam'):
  print (char)
```

```python
def merge(l, r):
  llen, rlen, i, j = len(l), len(r), 0, 0
  while i < llen or j < rlen:
    if j == rlen or (i < llen and l[i] < r[j]):
      yield l[i]
      i += 1
  else:
    yield r[j]
    j += 1
```

```python
g = merge([2,4], [1, 3, 5])
```

```python
while True:
    print (g.next())
```

```python
[x for x in merge([1,3,5],[2,4])]
```

```python
g = merge([2,4], [1, 3, 5])
```

```python
while True:
  try:
    print g.next()
  except StopIteration:
    print ('Done')
    break
```
